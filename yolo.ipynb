```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
model = YOLO('yolov8m.pt').to('cuda')  # load an official model
```

```python
image_np = cv2.imread('download.jpeg')
yolo_results = model(image_np, conf=0.4)
labeled_image = yolo_results[0].plot()
plt.figure()
plt.imshow(labeled_image[..., ::-1])  # change channels order since the YOLO work on BGR images
plt.show()
```

```python
results = model.train(data='config.yaml', epochs=50, batch=8)
```
