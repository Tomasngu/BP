```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
model = YOLO('runs/detect/train/weights/best.pt').to('cpu')  # load an official model

def show(img):
    plt.figure()
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # change channels order since the YOLO work on BGR images
    plt.show()
```

```python

yolo_results = model('cx/frame_8.jpg', conf=0.1)
labeled_image = yolo_results[0].plot()
plt.figure()
plt.axis('off')
plt.imshow(labeled_image[..., ::-1])  # change channels order since the YOLO work on BGR images
plt.show()
yolo_results[0].boxes
```
